BMI Calculator: https://www.nhlbi.nih.gov/calculate-your-bmi

BMI = (weight in pounds * 703) / (height in inches * height in inches)

BMI Index

<18.5 = Underweight

18.5 - 24.9 = Healthy

25.0 - 29.9 = Overweight

30.0 - 34.9 = Obese

35.0 - 39.9 = Severely Obese

=> 40.0 = Morbidly Obese

```python
#Variables
name = input("Enter your name: ")
weight = int(input("Enter your weight in pounds: "))
height_in_feet = int(input("Enter your height here in feet: "))
height_in_inches = int(input("Enter your height here in inches: "))

#Calculated variables
calculated_weight = weight * 703
total_height_in_inches = (height_in_feet * 12) + height_in_inches
BMI = calculated_weight/(total_height_in_inches * total_height_in_inches)

#Print results
print("Your BMI is: ", BMI)

#If statement
if BMI >0:
    if BMI <18.5:
        print(name + ", you are UNDERWEIGHT according to your BMI")
    elif BMI >=18.5 and BMI <=24.9:
        print(name + ", you are HEALTHY according to your BMI")
    elif BMI >=25.0 and BMI <=29.9:
        print(name + ", you are OVERWEIGHT according to your BMI")
    elif BMI >=30.0 and BMI <=34.9:
        print(name + ", you are OBESE according to your BMI")
    elif BMI >=35.0 and BMI <=39.9:
        print(name + ", you are SEVERELY OBESE according to your BMI")
    elif BMI >=40:
        print(name + ", you are MORBIDLY OBESE according to your BMI")
    else:
        print(name + ", please enter a valid inputs for weight and height")
```
