```python
#http://www.scrapethissite.com/pages/forms/

# import
from bs4 import BeautifulSoup
import requests

# variables
url = "http://www.scrapethissite.com/pages/forms/"
page = requests.get(url)
soup = BeautifulSoup(page.text, "html")
```

```python
soup.find("p", class_="lead")
```

```python
#Scraping a paragraph from the website

soup.find("p", class_="lead").text.strip()
```

```python
#All the table titles

soup.find_all("th")
```

```python
#All the table data

soup.find_all("td")
```

```python
#Only the Boston Bruins

soup.find("tr", class_="team").find("td", class_="name").text.strip()
```
