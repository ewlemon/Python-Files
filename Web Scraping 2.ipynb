```python
# Import

import pandas as pd
import requests
from io import StringIO

url = "https://en.wikipedia.org/wiki/List_of_largest_companies_by_revenue"

# Important variables
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"}
response = requests.get(url, headers=headers)
html = response.text
tables = pd.read_html(StringIO(html))

# First table
df = tables[0]

# Keep only the first 7 columns
df = df.iloc[:, :7]

# Rename columns
df.columns = ["Rank", "Name", "Industry", "Revenue", "Profit", "Employees", "Headquarters"]

# Clean revenue and profit columns
for col in ["Revenue", "Profit"]:
    df[col] = (
        df[col].astype(str)
        .str.replace(r"[^\d.-]", "", regex=True)
        .replace("", "0")
        .astype(float)
    )

# Clean employees column
df["Employees"] = (
    df["Employees"].astype(str)
    .str.replace(r"[^\d]", "", regex=True)
    .replace("", "0")
    .astype(int)
)

print(df.head())
```
