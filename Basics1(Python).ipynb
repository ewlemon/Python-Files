# While Loops

```python
number = 0

while number < 5:
    print(number)
    number = number + 1
```

```python
# Once number is 5 or more it will break the while loop, when it is false
```

```python
number = 0

while number < 5:
    print(number)
    if number == 3:
        break
    number = number + 1
```

```python
number = 0

while number < 5:
    print(number)
    if number == 6:
        break
    number = number + 1
else:
    print("No longer < 5")
```

```python
number = 0

while number < 5:
    number = number + 1
    if number == 3:
        continue
    print(number)
else:
    print("No longer < 5")
```

```python
def function():
    print("My first function")

function()
```

```python
def number_squared(number):
    print(number**2)

number_squared(4)
```

```python
def number_squared_cust(number, power):
    print(number**power)

number_squared_cust(4,2)
```

```python
args_tuple = (5,6,1,2,8)

def number_args(*number):
    print(number[0]*number[1])
```

```python
number_args(*args_tuple)
```

```python
# * is added to the def function to indicate that there is an unknown amount of variables, arbitrary argument
# it printed the first and second letter multiplied so 5*6
# it needs a * if a tuple is used in the function as well
```

```python
def number_squared_cust(number, power):
    print(number**power)

number_squared_cust(power = 4,number = 3)
```
