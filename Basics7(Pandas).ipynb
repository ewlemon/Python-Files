```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set index of df to "Date"

df = pd.read_csv(r"C:\Users\ewlem\Documents\Portfolio Projects\Python\Learning\Ice Cream Ratings.csv")
df= df.set_index("Date")
df
```

```python
# shows all styles for visualizations

print(plt.style.available)
```

```python
plt.style.use("bmh")
```

```python
# Default is line plot

df.plot()
```

```python
df.plot(kind="line", subplots=True)
```

```python
df.plot(kind="line", title="Ice Cream Ratings", xlabel="Daily Rating", ylabel="Scores")
```

```python
# Barplots can also be stacked

df.plot(kind="bar", stacked=True, title="Ice Cream Ratings", xlabel="Daily Rating", ylabel="Scores")
```

```python
# Only taking Flavor Rating

df["Flavor Rating"].plot(kind="bar")
```

```python
# barh is a horizontal bar

df.plot.barh(stacked=True)
```

```python
# scatter needs x and y axis

df.plot.scatter(x="Texture Rating", y="Overall Rating", s=100, c="Green")
```

```python
# Default bins are 10

df.plot.hist(bins=5)
```

```python
df.boxplot()
```

```python
# Figsize to increase or decrease size

df.plot.area(figsize=(10,5))
```

```python
# Pie charts need y

df.plot.pie(y="Flavor Rating", figsize=(10,6))
```
