```python
years = range(2000, 2012)
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.940, 0.950, 0.975]
oranges = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.895, 0.890, 0.888]
```

```python
plt.plot(years, apples, marker = "o")
plt.plot(years, oranges, marker = "x")
plt.xlabel("Year")
plt.ylabel("Yield (tons per hectare)")
plt.legend(["Apples", "Oranges"])
plt.title("Crop Yields in Kanto")
plt.show()
```

```python
# Matplotlib markers = https://matplotlib.org/stable/api/markers_api.html
```

```python
# plt. plot arguments
# color = c  https://matplotlib.org/stable/users/explain/colors/colors.html
# linestyle = ls (dashed or solid line) "-", "--", ""
# linewidth = lw (width of line) 1, 2, 3
# markersize = ms (size of markers) 1, 2, 3
# markeredgecolor = mec (color of markers)
# markeredgewidth = mew (width of markers) 1, 2, 3
# markerfacecolor = mfc (marker fill color) 
# alpha = aplha (opacity of plot) 0.5, 0.75
```

```python
plt.plot(years, apples, marker = "s", c="b", ls="-", lw=2, ms=8, mec="navy", mew=2)
plt.plot(years, oranges, marker = "o", c="r", ls="--", lw=3, ms=8, mew=6, alpha = 0.5)
plt.xlabel("Year")
plt.ylabel("Yield (tons per hectare)")
plt.legend(["Apples", "Oranges"])
plt.title("Crop Yields in Kanto")
plt.show()
```

```python
# FMT argument
# Apples is "s" shape, - line, and "b" color
# Oranges is "o" shape, -- line, and "r" color

plt.plot(years, apples, "s-b")
plt.plot(years, oranges, "o--r")
plt.xlabel("Year")
plt.ylabel("Yield (tons per hectare)")
plt.legend(["Apples", "Oranges"])
plt.title("Crop Yields in Kanto")
plt.show()
```

# Scatter plot

```python
# Import dataset from seaborn library

flowers_df = sns.load_dataset("iris")
```

```python
flowers_df
```

```python
flowers_df.species.unique()
```

```python
# Scatterplot needs an x and y value
# Hue to add a legend and differenciate, s for size

plt.figure(figsize=(12,6))
sns.scatterplot(x=flowers_df.sepal_length, y=flowers_df.sepal_width, hue=flowers_df.species, s=100)
plt.title("Sepal Dimensions")
plt.show()
```

```python
# Adding the data perimeter will allow you to reduce the amount of typing for x and y variables

sns.scatterplot(x="sepal_length", y="sepal_width", hue="species", s=100, data=flowers_df)
plt.title("Sepal Dimensions")
```

# Histogram

```python
flowers_df.sepal_width
```

```python
flowers_df.describe()
```

```python
# Bins changes the amount of bins/bars on the graph

plt.hist(flowers_df.sepal_width, bins=5)
plt.title("Distribution of Sepal Width")
plt.show()
```

```python
# Import numpy
# Arrange starts at 2 and goes to 5 in intervals of 0.25

import numpy as np

arange1= np.arange(2, 5, 0.25)
```

```python
plt.hist(flowers_df.sepal_width, bins=arange1)
plt.title("Distribution of Sepal Width")
plt.show()
```

```python
# Bins of unequal sizes

plt.hist(flowers_df.sepal_width, bins=[1,3,4,4.5])
plt.title("Distribution of Sepal Width")
plt.show()
```

```python
# Multiple Histograms

setosa_df = flowers_df[flowers_df.species == "setosa"]
versicolor_df = flowers_df[flowers_df.species == "versicolor"]
virginica_df = flowers_df[flowers_df.species == "virginica"]
```

```python
plt.hist(setosa_df.sepal_width, alpha=0.4, bins=arange1)
plt.hist(versicolor_df.sepal_width, alpha=0.4, bins=arange1)
plt.title("Distribution Based on Species")
plt.legend(["Setosa", "Versicolor"])
plt.show()
```

```python
plt.hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width], bins=arange1, stacked=True)
plt.title("Distribution Based on Species")
plt.legend(["Setosa", "Versicolor", "Virginica"])
plt.show()
```

# Heatmap

```python
flights_df = sns.load_dataset("flights")
flights_df
```

```python
plt.plot(flights_df.passengers)
plt.show()
```

```python
# Pivot Data
# Pivoting the data made month the index (rows), year the columns, and all the values became passengers
# Passengers are in thousands

pivoted_df = flights_df.pivot(index="month", columns="year", values="passengers")
pivoted_df
```

```python
sns.heatmap(pivoted_df)
plt.title("No. of passengers (1000s)")
plt.show()
```

```python
# fmt (format) for the numbers
# annot (annotation) adds numbers to the heatmap
# cmap (color maps) changes the color

sns.heatmap(pivoted_df, fmt="d", annot=True, cmap="Blues")
plt.title("No. of passengers (1000s)")
plt.show()
```

# Images

```python
# Import

import matplotlib.pyplot as plt
from urllib.request import Request, urlopen
from PIL import Image
from io import BytesIO
```

```python
# Set image url as variable
# The req area of code is to bypass internet security in site
# Opening image using BytesIO

image_url = "https://i.imgur.com/SkPbq.jpeg"

req = Request(image_url, headers={"User-Agent": "Mozilla/5.0"})
with urlopen(req) as response:
    img_data = response.read()

img = Image.open(BytesIO(img_data))

plt.imshow(img)
plt.grid(False)
plt.axis("off")
plt.title("A Data Science Meme")
plt.show()
```

# Plotting multiple charts in a grid

```python
# Example of subplots

plt.subplots(2,3)
plt.tight_layout(pad=2)
plt.show()
```

```python
# Variables

years = range(2000, 2012)
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.937, 0.940, 0.950, 0.975]
oranges = [0.962, 0.941, 0.930, 0.923, 0.918, 0.908, 0.907, 0.904, 0.901, 0.895, 0.890, 0.888]
```

```python
# Adding charts to subplots
# Adjusting size of charts
# plt.tight_layout adds a padding to the graphs

fig, axes = plt.subplots(2, 3, figsize=(16,8))
plt.tight_layout(pad=3)

# Each chart has an index/axes
# Axes[0,0] is the first graph (Others are [0,1], [0,2], [1,0], [1,1], [1,2])
# Note that set_title, set_xlabel, set_ylabel is used instead of title to avoid error

axes[0,0].plot(years, oranges, "o--r")
axes[0,0].plot(years, apples, "s-b")
axes[0,0].set_title("Yield of Oranges")
axes[0,0].set_xlabel("Year")
axes[0,0].set_ylabel("Yield (tons per hectare)")
axes[0,0].legend(["Apples", "Oranges"])

# When using sns or seaborn graphs they can't follow axes[] and must be added in as ax=axes[]

axes[0,1].set_title("Sepal length vs width")
sns.scatterplot(x= flowers_df.sepal_length, y=flowers_df.sepal_width, hue=flowers_df.species, s=100, ax=axes[0,1])

axes[0,2].set_title("Distribution of Sepal Width")
axes[0,2].hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width], bins=arange1, stacked=True)
axes[0,2].legend(["Setosa", "Versicolor", "Virginica"])

axes[1,0].set_title("Restaurant Bills")
sns.barplot(x="day", y="total_bill", hue="sex", data=tips_df, ax=axes[1,0])

axes[1,1].set_title("Flight Traffic")
sns.heatmap(pivoted_df, cmap="Blues", ax=axes[1,1])

axes[1,2].set_title("Data Science Meme")
axes[1,2].imshow(img)
axes[1,2].grid(False)
# Hiding grid and x+y axes ticks
axes[1,2].set_xticks([])
axes[1,2].set_yticks([])

plt.show()
```
