```python
# Merge: outer, inner, left, right

df1.merge(df2)

# Default is inner join, only same from both tables will be shown
```

```python
df1.merge(df2, how="inner", on="FellowshipID")

# Only joining on fellowshipid will not join the names together
```

```python
df1.merge(df2, how="outer")

# Pulls in all values from both tables
# No data for age and skills on some
```

```python
df1.merge(df2, how="left")

# Pulls all data from left and matching data from right
```

```python
df1.merge(df2, how="right")

# Pulls all data from right and matching data from left
```

```python
df1.merge(df2, how="cross")

# Pulls all data from left and compares each line in right to each row in left
```

```python
df1.join(df2, on="FellowshipID", how="outer", lsuffix="_left", rsuffix="_right")
```

```python
# Join is best with index
# Setting FellowshipID as index

df3 = df1.set_index("FellowshipID").join(df2.set_index("FellowshipID"), lsuffix="_left", rsuffix="_right", how="outer")
```

```python
# concat() puts one table on top of the other and can join as well

pd.concat([df1,df2], join="outer", axis = 1)
```
